Skrip Koleksi Data Mentah (Scraper)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# 1. Tentukan daftar URL sumber data (Ganti dengan URL target Anda)
urls = [
    "https://contoh-situs-kesehatan.com/penyakit/influenza",
    "https://contoh-situs-kesehatan.com/penyakit/diabetes"
]

# List untuk menampung data mentah
raw_dataset = []

for url in urls:
    try:
        # Mengirim permintaan ke website
        headers = {'User-Agent': 'Mozilla/5.0'}
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        
        # Parsing HTML
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # 2. Ekstraksi elemen (SESUAIKAN TAG DAN CLASS DENGAN WEBSITE TARGET)
        # Ekstraksi dilakukan secara mentah (raw text) tanpa pembersihan
        nama_penyakit = soup.find('h1', class_='judul-penyakit')
        deskripsi = soup.find('div', class_='konten-deskripsi')
        gejala = soup.find('div', class_='konten-gejala')
        pengobatan = soup.find('div', class_='konten-pengobatan')
        jenis_obat = soup.find('div', class_='konten-jenis-obat')
        dosis_obat = soup.find('div', class_='konten-dosis')
        
        # Memasukkan data ke dalam dictionary
        # Menggunakan .get_text() hanya untuk mengambil teks dari HTML, tanpa strip() atau lower()
        raw_dataset.append({
            "Nama_Penyakit": nama_penyakit.get_text() if nama_penyakit else "",
            "Deskripsi_Penyakit": deskripsi.get_text() if deskripsi else "",
            "Gejala": gejala.get_text() if gejala else "",
            "Cara_Pengobatan": pengobatan.get_text() if pengobatan else "",
            "Jenis_Obat": jenis_obat.get_text() if jenis_obat else "",
            "Dosis_Obat": dosis_obat.get_text() if dosis_obat else "",
            "URL_Sumber": url
        })
        
        print(f"Berhasil mengambil data dari: {url}")
        
        # Jeda waktu agar tidak terblokir oleh server target (Rate limiting)
        time.sleep(2)
        
    except Exception as e:
        print(f"Gagal mengambil data dari {url} | Error: {e}")

# 3. Konversi ke DataFrame
df_raw = pd.DataFrame(raw_dataset)

# 4. Simpan ke CSV tanpa preprocessing tambahan
file_name = "data\raw_dataset_medis_rag.csv"
df_raw.to_csv(file_name, index=False, encoding='utf-8')

print(f"\nSelesai! Data mentah berhasil disimpan ke {file_name}")
# Menampilkan 5 baris pertama sebagai preview
df_raw.head()
```
